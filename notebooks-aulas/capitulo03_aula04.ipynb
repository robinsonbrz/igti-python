### Conversão dos tipos de Dados

```python
nascimento = 1970
ano_atual = 2022
idade = ano_atual - nascimento
saida = 'Eu tenho ' + str(idade) + ' anos'
print(saida)
```

```python
#conversão do tipo int
print(float(10))
print(bool(-1))
print(bool(0))
print(str(-999))
```

```python
#conversão do tipo float
print(int(9.999))
print(bool(-0.99))
print(str(-0.99))
```

```python
#conversão do tipo bool
print(int(True))
print(float(True))
print(str(False))
```

```python
#conversão do tipo str
print(int('-99')) # -99
print(float('0.01')) # 0.01
print(bool('palavra')) # True
print(bool('')) # False
```

```python
#particularidades com o tipo None
print(str(None))
print(bool(None))
```

### Formatação utilizando f-strings



```python
variavel = 99
f'Exemplo de f-string com um valor {variavel} e uma expressão {variavel + 1}'
```

```python
nascimento = 1986
ano_atual = 2022
idade = ano_atual - nascimento
print(f'Eu tenho {idade} anos')
```

```python
nascimento = 1986
ano_atual = 2022
print(f'Eu tenho {ano_atual - nascimento} anos')
```

```python
palavra = 'ave'
print(f'A palavra {palavra.upper()} possuí {len(palavra)} letras')
```

```python
orcamento = 3
vlr_gasto = 10
pct = (vlr_gasto/orcamento) * 100
print(f'Porcentagem já gasta do orçamento: {round(pct,2)}%')
```

```python
orcamento = 3
vlr_gasto = 11
pct = (vlr_gasto/orcamento) * 100
print(f'Porcentagem já gasta do orçamento: {pct:.20f}%')
```

```python
orcamento = 5000
vlr_gasto = 430
pct = (vlr_gasto/orcamento)
print(f'Porcentagem já gasta do orçamento: {pct:.5%}')
```
